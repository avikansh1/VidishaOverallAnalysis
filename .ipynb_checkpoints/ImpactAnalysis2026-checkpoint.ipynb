```python
import pandas as pd
import numpy as np

# -----------------------------
# 1. Read file
# -----------------------------
file_path = r"C:\Users\Avikansh\Downloads\2026_Analysis\Sironj Vidisha Baseline Assessment Jan 2026 Analysis Calculator.xlsx"

df = pd.read_excel(file_path, header=1, sheet_name='Data')

# -----------------------------
# 2. Check original columns
# -----------------------------
print("Total columns:", len(df.columns))
print("Original columns:")
print(df.columns.tolist())

# -----------------------------
# 3. Convert all column names to string
#    so 1,2,3...22 become '1','2','3'...
# -----------------------------
df.columns = [str(col).strip() for col in df.columns]

# -----------------------------
# 4. Rename question columns
#    Fixed columns = first 14
#    Question columns:
#    1-7   -> Hindi Reading
#    8-11  -> Hindi Writing
#    12-22 -> Numeracy
# -----------------------------
fixed_cols = list(df.columns[:14])

reading_cols = [f'Hindi_Reading_{i}' for i in range(1, 8)]     # 7 columns
writing_cols = [f'Hindi_Writing_{i}' for i in range(1, 5)]     # 4 columns
numeracy_cols = [f'Numeracy_Q{i}' for i in range(1, 12)]       # 11 columns

new_cols = fixed_cols + reading_cols + writing_cols + numeracy_cols

print("\nNew column count:", len(new_cols))

df.columns = new_cols

# -----------------------------
# 5. Check renamed columns
# -----------------------------
print("\nRenamed columns:",df.columns.tolist())

# -----------------------------
# 6. Convert question columns to numeric
# -----------------------------
question_cols = reading_cols + writing_cols + numeracy_cols

for col in question_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# -----------------------------
# 7. Recalculate subject totals from question columns
# -----------------------------
df['Reading_Calculated'] = df[reading_cols].sum(axis=1)
df['Writing_Calculated'] = df[writing_cols].sum(axis=1)
df['Numeracy_Calculated'] = df[numeracy_cols].sum(axis=1)

# -----------------------------
# 8. Convert original total columns to numeric
# -----------------------------
df['Reading'] = pd.to_numeric(df['Reading'], errors='coerce')
df['Writing'] = pd.to_numeric(df['Writing'], errors='coerce')
df['Numeracy'] = pd.to_numeric(df['Numeracy'], errors='coerce')

# -----------------------------
# 9. Compare original vs calculated totals
# -----------------------------
df['Reading_Match'] = np.where(df['Reading'] == df['Reading_Calculated'], 'Match', 'Mismatch')
df['Writing_Match'] = np.where(df['Writing'] == df['Writing_Calculated'], 'Match', 'Mismatch')
df['Numeracy_Match'] = np.where(df['Numeracy'] == df['Numeracy_Calculated'], 'Match', 'Mismatch')

# -----------------------------
# 10. Show match summary
# -----------------------------
print("\nReading Match Summary:")
print(df['Reading_Match'].value_counts(dropna=False))

print("\nWriting Match Summary:")
print(df['Writing_Match'].value_counts(dropna=False))

print("\nNumeracy Match Summary:")
print(df['Numeracy_Match'].value_counts(dropna=False))

# -----------------------------
# 11. Optional: create percentages from calculated totals
#     Only do this if you want your own calculated %
#     Update denominators if max marks are different
# -----------------------------
reading_max = 27   # change if needed
writing_max = 18   # change if needed
numeracy_max = 30  # change if needed

df['Reading_Calc_%'] = (df['Reading_Calculated'] / reading_max) * 100
df['Writing_Calc_%'] = (df['Writing_Calculated'] / writing_max) * 100
df['Numeracy_Calc_%'] = (df['Numeracy_Calculated'] / numeracy_max) * 100

# -----------------------------
# 12. Save cleaned file
# -----------------------------
output_path = r"C:\Users\Avikansh\Downloads\2026_Analysis\Sironj_Baseline_Cleaned.xlsx"
df.to_excel(output_path, index=False)

print("\nCleaned file saved successfully at:")
print(output_path)

# -----------------------------
# 13. Preview data
# -----------------------------
print("\nPreview:")
print(df.head())
```
