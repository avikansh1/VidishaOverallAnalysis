```python
def standardize_df(df):
    df = df.copy()
    
    # clean column names
    df.columns = df.columns.str.strip().str.replace(" ", "_")
    
    # clean text columns if present
    text_cols = ['Student_Name', 'School_Name', 'Group', 'Grade', 'Gender_(M/F)']
    for col in text_cols:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip()
    
    # standardize group names
    if 'Group' in df.columns:
        mapping = {
            'treatment': 'Treatment',
            'intervention': 'Treatment',
            'control': 'Control',
            'reference': 'Control',
            'general': 'General',
            'demo': 'General'
        }
        df['Group'] = df['Group'].str.lower().map(mapping).fillna(df['Group'])
    
    # convert score columns to numeric if present
    score_cols = ['Reading_Calculated', 'Writing_Calculated', 'Numeracy_Calculated']
    for col in score_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
    
    return df

baseline_df = standardize_df(baseline_df)
midline_df = standardize_df(midline_df)

baseline_df.head()
```

```python
print("Baseline columns:\n", baseline_df.columns.tolist())
print("\nMidline columns:\n", midline_df.columns.tolist())
```

```python
score_cols = ['Reading_Calculated', 'Writing_Calculated', 'Numeracy_Calculated']

baseline_df = baseline_df.dropna(subset=score_cols).copy()
midline_df = midline_df.dropna(subset=score_cols).copy()

print("Baseline after score filtering:", baseline_df.shape)
print("Midline after score filtering:", midline_df.shape)
```

```python
# Create merge ID

for d in [baseline_df, midline_df]:
    d['Student_Name'] = d['Student_Name'].astype(str).str.strip().str.lower()
    d['School_Name'] = d['School_Name'].astype(str).str.strip().str.lower()
    d['Grade'] = d['Grade'].astype(str).str.strip()

    d['id'] = (
        d['Student_Name'] + "_" +
        d['School_Name'] + "_" +
        d['Grade']
    )
```

```python
# Remove duplicates before merge

baseline_match = baseline_df.drop_duplicates(subset='id').copy()
midline_match = midline_df.drop_duplicates(subset='id').copy()

print("Unique baseline IDs:", baseline_match['id'].nunique())
print("Unique midline IDs:", midline_match['id'].nunique())
```
